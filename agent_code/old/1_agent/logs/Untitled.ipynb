```python
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

score=[]
with open("Score1.dat") as f:
    for line in f:
        score.append(float(line))
        
num_bins = 299
# the histogram of the data
n, bins, patches = plt.hist(score, num_bins, range=[0,9],histtype='bar',facecolor='green')
plt.xlabel(r'$Score$')
plt.ylabel(r'$P(Score)$')
plt.show()
```
